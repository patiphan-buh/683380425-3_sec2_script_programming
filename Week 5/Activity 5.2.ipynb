กิจกรรมที่ 5.2: โปรแกรมนับความถี่ของคำ (Word Frequency Counter) โดยใช้ Dictionary

```python
# 🔧 เติมโค้ด: Activity 5.2 - Word Frequency Counter
# ------------------------------------------------------------
# โจทย์: เขียนโปรแกรมนับจำนวนความถี่ของแต่ละคำในประโยค

# --- ข้อที่ 1: รับข้อความจากผู้ใช้ ---
sentence = input("Enter a sentence: ")

# todo (1): แปลงข้อความเป็นตัวพิมพ์เล็ก
sentence = sentence.lower()

# todo (2): แยกข้อความออกเป็นคำ
words = sentence.split()

# --- ข้อที่ 2: สร้าง Dictionary เปล่าสำหรับนับคำ ---
word_counts = {}

# --- ข้อที่ 3: วนลูปนับจำนวนคำ ---
for word in words:

    # Challenge: ลบเครื่องหมายวรรคตอนออกจากแต่ละคำ
    # todo (3): ใช้ strip() เพื่อลบ .,!?"
    word = word.strip('.,!?"')

    # todo (4): ตรวจสอบว่าคำนี้มีอยู่ใน Dictionary แล้วหรือยัง
    if word in word_counts:
        word_counts[word] += 1
    else:
        # todo (5): เพิ่มคำใหม่พร้อมกำหนดค่าเริ่มต้นเป็น 1
        word_counts[word] = 1

# --- ข้อที่ 4: แสดงผลลัพธ์ ---
print("\nWord Frequency")
for word, count in word_counts.items():
    print(f"{word}: {count}")
```

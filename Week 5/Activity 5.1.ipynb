กิจกรรมที่ 5.1: การประยุกต์ใช้ Dictionary และ Set ในชีวิตจริง

คุณจะนำ Dictionary ไปใช้ตรงไหนในเกมได้บ้าง?

ตอบ: ใช้เก็บข้อมูลผู้เล่น เช่น ช่องเก็บของ (Inventory), ค่าพลัง (HP, MP, Attack, Defense), ข้อมูลตัวละคร หรือข้อมูลภารกิจ (Quest) เพราะ Dictionary เก็บข้อมูลแบบ Key–Value ทำให้ค้นหาและแก้ไขข้อมูลได้รวดเร็ว

Dictionary จะช่วยจัดเก็บข้อมูลสินค้าในร้านค้าออนไลน์ได้อย่างไร?

ตอบ: ใช้รหัสสินค้า (Product ID) เป็น Key และเก็บรายละเอียดสินค้า เช่น ชื่อ ราคา จำนวนคงเหลือ หมวดหมู่ หรือยี่ห้อ เป็น Value ทำให้ค้นหาข้อมูลสินค้าได้สะดวก

Set จะมีประโยชน์อย่างไรในการจัดการสิทธิ์ของผู้ใช้ หรือการจัดการแท็ก?
 
ตอบ: ใช้เก็บสิทธิ์ผู้ใช้ (Admin, Editor, Viewer), แฮชแท็กที่ไม่ซ้ำกัน, รายชื่อสมาชิกชมรม หรือรายชื่อผู้เข้าร่วมกิจกรรม เพราะ Set ไม่อนุญาตให้มีข้อมูลซ้ำและตรวจสอบการมีอยู่ของข้อมูลได้อย่างรวดเร็ว

มีสถานการณ์ใดบ้างที่ต้องตรวจสอบอย่างรวดเร็วว่า "ข้อมูลนี้มีอยู่หรือไม่" โดยไม่สนใจลำดับของข้อมูล?
 
 ตอบ: เช่น ตรวจสอบ IP Address ที่ถูกบล็อก, อีเมลในบัญชีดำ, รายชื่อผู้เข้าชมเว็บไซต์แบบไม่ซ้ำ, รหัสนักศึกษาที่ลงทะเบียนแล้ว, หมายเลขสมาชิก หรือรหัสคูปองที่เคยใช้งานแล้ว เพราะ Set เหมาะสำหรับการค้นหาข้อมูลและป้องกันข้อมูลซ้ำได้อย่างมีประสิทธิภาพ

Activity 4.2: Interactive To-Do List (using List Operations)

```python
print("=== To-Do List ===")

todo_list = []

while True:
    print("\n===== MENU =====")
    print("1. Add a task")
    print("2. View tasks")
    print("3. Complete a task")
    print("4. Exit")

    choice = input("Choose an option (1-4): ")

    # 1. เพิ่มงาน
    if choice == "1":
        task = input("Enter a new task: ")
        todo_list.append(task)
        print(f"'{task}' has been added.")

    # 2. แสดงรายการงานทั้งหมด
    elif choice == "2":
        if len(todo_list) == 0:
            print("No tasks in the list.")
        else:
            print("\nYour To-Do List:")
            for i, task in enumerate(todo_list):
                print(f"{i}. {task}")

    # 3. ทำงานเสร็จ (ลบออก)
    elif choice == "3":
        if len(todo_list) == 0:
            print("No tasks to remove.")
        else:
            print("\nCurrent Tasks:")
            for i, task in enumerate(todo_list):
                print(f"{i}. {task}")

            remove_choice = input(
                "Remove by (1) Index or (2) Task Name? ")

            if remove_choice == "1":
                index = int(input("Enter task index: "))
                if 0 <= index < len(todo_list):
                    removed = todo_list.pop(index)
                    print(f"'{removed}' has been completed.")
                else:
                    print("Invalid index.")

            elif remove_choice == "2":
                task_name = input("Enter task name: ")
                if task_name in todo_list:
                    todo_list.remove(task_name)
                    print(f"'{task_name}' has been completed.")
                else:
                    print("Task not found.")

            else:
                print("Invalid choice.")

    # 4. ออกจากโปรแกรม
    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid menu. Please choose 1-4.")
```

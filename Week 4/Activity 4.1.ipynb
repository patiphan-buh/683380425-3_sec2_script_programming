Activity 4.1: List vs. Tuple: Scenario Matching

1. ข้อมูลอุณหภูมิประจำวันที่มีการบันทึกไว้ตลอดหนึ่งสัปดาห์

ตอบ = ควรใช้ List เพราะ ข้อมูลอาจมีการแก้ไขหรือบันทึกค่าใหม่ภายหลังได้ จึงควรใช้ List ที่สามารถเปลี่ยนแปลงข้อมูลได้

2. ค่าสี RGB ของเฉดสีเฉพาะเฉดหนึ่ง (เช่น (255, 0, 0))

ตอบ = ควรใช้ Tuple เพราะ ค่าสี RGB ของสีหนึ่งๆ เป็นค่าคงที่ ไม่ควรเปลี่ยนแปลงหลังจากกำหนด

3. รายการสินค้าในรถเข็นชอปปิงที่สามารถกดเพิ่มหรือลบออกได้

ตอบ = ควรใช้ List เพราะ รายการสินค้าเปลี่ยนแปลงได้ตลอดเวลา สามารถเพิ่ม ลบ หรือแก้ไขข้อมูลได้

4. รายชื่อวันในหนึ่งสัปดาห์เรียงตามลำดับ (จันทร์ - อาทิตย์)

ตอบ = ควรใช้ Tuple เพราะ วันในสัปดาห์เป็นข้อมูลคงที่ไม่เปลี่ยนแปลงจึงเหมาะกับ Tuple

5. คะแนนเก็บของนักเรียนจากการส่งงานชุดต่างๆ ซึ่งอาจมีการอัปเดตหรือแก้ไขคะแนนในภายหลังได้

ตอบ = ควรใช้ List เพราะ คะแนนสามารถแก้ไขหรืออัปเดตได้เมื่อมีการตรวจใหม่หรือให้คะแนนเพิ่มเติม

6. พิกัด GPS ของสถานที่สำคัญหรือจุดแลนด์มาร์กที่ตั้งอยู่กับที่

ตอบ = ควรใช้ Tuple เพราะ พิกัดของสถานที่เป็นข้อมูลคงที่ไม่ควรแก้ไขจึงเหมาะกับ Tuple

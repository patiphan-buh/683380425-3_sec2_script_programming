Activity 3.2: Pattern Printing with Nested Loops

```python
# Square Pattern
rows = 5

for i in range(rows):
    for j in range(rows):
        print("*", end="")
    print()
```

```python
# Right-Angled Triangle
rows = 5

for i in range(1, rows + 1):
    for j in range(i):
        print("*", end="")
    print()
```

```python
# Inverted Triangle
rows = 5

for i in range(rows, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```

```python
# Challenge: Pyramid
rows = 5

for i in range(rows):
    for j in range(rows - i - 1):
        print(" ", end="")
    for k in range(2 * i + 1):
        print("*", end="")
    print()
```

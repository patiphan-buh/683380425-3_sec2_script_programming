1. สคริปต์สั้นๆ Word Problem

```python
celsius = float(input("ป้อนอุณหภูมิ (°C): "))

fahrenheit = (celsius * 9/5) + 32


print(f"{celsius}°C แปลงเป็นฟาเรนไฮต์ได้: {fahrenheit}°F")
```

ออกแบบเกมสั้น ๆ แบบ CHOOSE YOUR OWN ADVENTURE Tay กําหนดเนื้อเรื่อง จุดตัดสินใจ (DECISION) พร้อมเขียน ลําดับการทํางานด้วย INPUT() และ IF-ELIF- ELSE (ไม่จําเป็นต้องเขียนโค้ด PYTHON)
และผลลัพธ์ของแต่ละทางเลือก

```python
print("=== CHOOSE YOUR OWN ADVENTURE ===")
print("You wake up in a dark forest...")

choice1 = input("Do you go LEFT or RIGHT? ").lower()

if choice1 == "left":
    print("You walk left and find a river.")

    choice2 = input("Do you SWIM across or FOLLOW the river? ").lower()

    if choice2 == "swim":
        print("You try to swim... and find a treasure chest! ")

    elif choice2 == "follow":
        print("You follow the river and find a village")

    else:
        print("You got lost in the forest")

elif choice1 == "right":
    print("You walk right and see a cave.")

    choice2 = input("Do you ENTER or RUN away? ").lower()

    if choice2 == "enter":
        print("A dragon appears! But it lets you pass.")

    elif choice2 == "run":
        print("You safely escape, but miss an adventure.")

    else:
        print("You stand there confused until night falls ")

else:
    print("You stand still and the adventure ends...")
```

เกมเริ่มต้นจากการแสดงข้อความต้อนรับและบอกเนื้อเรื่องว่าผู้เล่นตื่นขึ้นมาในป่ามืด จากนั้นโปรแกรมจะรับข้อมูลจากผู้เล่นด้วย input() เพื่อให้เลือกว่าจะไปทาง LEFT หรือ RIGHT และนำค่าที่ได้มาแปลงเป็นตัวพิมพ์เล็กด้วย .lower() เพื่อช่วยให้เปรียบเทียบเงื่อนไขได้ง่ายขึ้น

เมื่อผู้เล่นเลือก “left” โปรแกรมจะเข้าเงื่อนไข if และแสดงว่าเจอแม่น้ำ จากนั้นจะมีการตัดสินใจอีกขั้นโดยใช้ input() อีกครั้งเพื่อเลือก SWIM หรือ FOLLOW แล้วใช้ if-elif-else ตรวจสอบผลลัพธ์ทีละเงื่อนไข หากเลือก swim จะเจอสมบัติ หากเลือก follow จะเจอหมู่บ้าน แต่ถ้าไม่ตรงกับทั้งสองตัวเลือกจะเข้าสู่ else และหลงป่า

แต่ถ้าผู้เล่นเลือก “right” โปรแกรมจะเข้า elif และแสดงว่าเจอถ้ำ จากนั้นมีการตัดสินใจอีกขั้นเหมือนกัน โดยให้เลือก ENTER หรือ RUN แล้วใช้ if-elif-else ตรวจสอบ ถ้าเลือก enter จะเจอมังกรแต่รอดได้ ถ้าเลือก run จะหนีได้อย่างปลอดภัย แต่ถ้าไม่ตรงเงื่อนไขจะเข้าสู่ else และติดอยู่ในความมืด

สุดท้ายถ้าผู้เล่นพิมพ์ค่าอื่นที่ไม่ใช่ left หรือ right โปรแกรมจะเข้าสู่ else หลัก และจบเกมทันทีโดยไม่มีการผจญภัยต่อ

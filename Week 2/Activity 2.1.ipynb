ให้อธิบายลําดับการทํางานของโปรแกรมที่ใช้ IF,ELIF,ELSE และ LOGICAL OPERATORS (AND,OR,NOT) โดยพิจารณาเงื่อนไขทีละ ขั้นตอน และบอกผลลัพธ์ (OUTPUT) ที่โปรแกรมจะแสดง สําหรับค่าที่ กําหนดให้

Snippet A

```python
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
else:
    grade = "C"

print(f"Grade: {grade}")
```

Ans: โปรแกรมเริ่มจากกำหนดค่า score = 85 แล้วตรวจสอบเงื่อนไข if score >= 90 ซึ่งเป็นเท็จจึงไม่ทำงาน จากนั้นไปตรวจ elif score >= 80 ซึ่งเป็นจริงเพราะ 85 มากกว่าหรือเท่ากับ 80 ทำให้กำหนดค่า grade = "B" และข้าม else ไป
Output
Grade: B

Snippet B

```python
temp = 25
is_raining = True

if temp > 20 and not is_raining:
    print("Perfect day for a picnic!")
elif temp <= 20 or is_raining:
    print("Stay indoors or bring an umbrella.")
else:
    print("Unsure about the weather.")
```

Ans: โปรแกรมกำหนดค่า temp = 25 และ is_raining = True จากนั้นตรวจสอบเงื่อนไขแรก temp > 20 and not is_raining โดยพบว่า temp > 20 เป็นจริง แต่ not is_raining เป็นเท็จเพราะ is_raining เป็น True ทำให้เงื่อนไข AND เป็น False จึงไม่เข้า if ต่อไปโปรแกรมตรวจ elif temp <= 20 or is_raining ซึ่งพบว่า is_raining เป็น True ทำให้เงื่อนไข OR เป็นจริง

Output: Stay indoors or bring an umbrella


Snippet C

```python
age = 17
has_license = True

if age >= 18:
    if has_license:
        print("You can drive legally.")
    else:
        print("You need a license to drive.")
else:
    print("You are too young to drive.")
```

Ans: โปรแกรมกำหนด age = 17 และ has_license = True จากนั้นตรวจสอบเงื่อนไข if age >= 18 ซึ่งเป็นเท็จเพราะอายุยังไม่ถึง 18 ทำให้ไม่เข้าเงื่อนไขด้านใน และโปรแกรมไปทำงานในส่วน else  
Output: You are too young to drive
